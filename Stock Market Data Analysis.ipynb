Daily Percentage Change

```python
tesla['returns'] = (tesla['Close']/tesla['Close'].shift(1)) - 1
```

```python
tesla.head()
```

```python
ford['returns'] = (ford['Close']/ford['Close'].shift(1)) - 1
```

```python
ford.head()
```

```python
gm['returns'] = (gm['Close']/gm['Close'].shift(1)) - 1
```

```python
gm.head()
```

Let's now plot the volatility of the graph

```python
ford['returns'].hist(bins=50)
```

The higher the standard deviation is , the more the variance is seen

```python
tesla['returns'].hist(bins=100 , label = 'Tesla',alpha=0.5, figsize = (15,6))
gm['returns'].hist(bins = 100, label = 'GM' , alpha = 0.5)
ford['returns'].hist(bins=100, label='Ford',alpha =0.5)
plt.legend()
```

Cumulative Return

Date           Price
01/01/2020       10
01/02/2020       15
01/03/2020       20
01/03/2020       25

Date           Daily Return     %Daily Return
01/01/2020                          -
01/02/2020       15/10             50%
01/03/2020       20/15             33% 
01/03/2020       25/20             20%

```python
tesla['Cumulative Return'] = (1+tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cumulative Return'] = (1+ford['returns']).cumprod()
gm['Cumulative Return'] = (1+gm['returns']).cumprod()
tesla['Cumulative Return'] = (1+tesla['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='Tesla',figsize=(15,7))
gm['Cumulative Return'].plot(label='GM')
ford['Cumulative Return'].plot(label = 'Ford')
plt.title('Cumulative Return vs time')
plt.legend()
```

In the above graph we can assume that tesla has a higher rate of retuen when compared to GM , Ford.
when you did invest 1 $ in 2012 in tesla , now in the year 2021 , it is nearly worth 100$
similarly ford and gm have the share price , thus they are stable stocks
